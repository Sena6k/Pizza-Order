```python
import time
from datetime import datetime
import pandas as pd
import csv
import logging

# Logger Class
#------------------------------------------------------------------------------

class CustomFormatter(logging.Formatter):

    grey = "\x1b[38;20m"
    yellow = "\x1b[33;20m"
    red = "\x1b[31;20m"
    bold_red = "\x1b[31;1m"
    reset = "\x1b[0m"
    format = "%(asctime)s - %(name)s - %(levelname)s - %(message)s (%(filename)s:%(lineno)d)"

    FORMATS = {
        logging.DEBUG: grey + format + reset,
        logging.INFO: grey + format + reset,
        logging.WARNING: yellow + format + reset,
        logging.ERROR: red + format + reset,
        logging.CRITICAL: bold_red + format + reset
    }

    def format(self, record):
        log_fmt = self.FORMATS.get(record.levelno)
        formatter = logging.Formatter(log_fmt)
        return formatter.format(record)



#------------------------------------------------------------------------------



class Pizza(object):
    def __init__(self,name_pizza,cost_pizza=0,description_pizza=[],**kwargs):
      self.cost_pizza = cost_pizza
      self.description_pizza = description_pizza
      self.__name_pizza = name_pizza

      if name_pizza == "Margherita":
        self.__name_pizza = "Margherita"
        self.description_pizza =["Mozarella","Pizza Sauce"]
        self.cost_pizza = 25

      elif name_pizza == "Superos":
        self.description_pizza = ["Mozarella","PizzaSauce","Mushroom","Corn","Sausage","Olive"]
        self.cost_pizza = 45
        self.__name_pizza = "Superos"

      elif name_pizza == "Classic":
        self.description_pizza = ["Mozarella","PizzaSauce","Mushroom","Tomato","Corn"]
        self.cost_pizza = 30
        self.__name_pizza = "Classic"

      elif name_pizza == "Plain":
        self.description_pizza = ["Mozarella","PizzaSauce","Corn"]
        self.cost_pizza = 40
        self.__name_pizza = "Plain"

      elif name_pizza == "Turkish":
        self.description_pizza = ["Mozarella","PizzaSauce","Mushroom","Tomato","Sausage"]
        self.cost_pizza = 35
        self.__name_pizza == "Turkish"

    def __add__(self,add_items):

       self.add_items = add_items
       return Pizza.get_description.append(add_items)


    def __add__(self,add):
       self.add = add
       return Pizza.get_cost(add)

    def get_description(self):
        return self.description_pizza
    def get_cost(self):
        return self.cost_pizza

    def display_pizza(self):
      return f'-----> Your pizza including: ---> {self.description_pizza} and total cost:---> {self.cost_pizza}'


    # pizza describes



#-------------------------------------

class Sauces(object):
    def __init__(self,Sauces_data={},**kwargs):

        self.Sauces_data = {'Olives': 5,
                            'Mushroom': 5,
                            'Goat cheese': 8,
                            'Meat': 8,
                            'Onion': 5,
                            'Corn': 6}

    def getSauces(self):
        return self.Sauces_data



class AddToPizza(Sauces,Pizza):
    def __init__(self,Sauces,__name_pizza,cost_pizza,description_pizza,add,adding_cost, **kwargs):
        print("Add class")
        self.add = add
        self.Sauces=Sauces
        self.Pizza=Pizza
        self.cost_pizza = cost_pizza
        self.description_pizza = description_pizza
        self.adding_cost=adding_cost


        #super(Pizza, self).__init__(__name_pizza, cost_pizza, description_pizza)
        #super(Sauces, self).__init__(Sauces.Sauces_data)
        #super().__init__(__name_pizza, cost_pizza, description_pizza)
        Pizza.__init__(self, __name_pizza, cost_pizza, description_pizza,**kwargs)

    def definition_new(self):
        self.description_pizza = self.description_pizza + self.add
        return self.description_pizza
    def cost_new(self):
        self.cost_pizza = self.adding_cost + self.cost_pizza
        return self.cost_pizza
    def display(self):
       return(f'Your pizza contents: {self.description_pizza} and total cost: {self.cost_pizza}')

#--------------------------------------------------------------------------------------------------

# create logger with 'spam_application'
logger = logging.getLogger("Pizza Order")
logger.setLevel(logging.DEBUG)

# create console handler with a higher log level
ch = logging.StreamHandler()
ch.setLevel(logging.DEBUG)

ch.setFormatter(CustomFormatter())

logger.addHandler(ch)



#Menu open..

file_menu = open(r'Menu.txt')
print(file_menu.read())


# ----------------------------------


# we created sauces object

sauces = Sauces()

# addition input

list_pizza_name = ["Superos","Turkish","Plain","Margherita","Classic"]
x = str(input("What would you like to order ? (Just you write pizza name) :  "))
assert x.capitalize() in list_pizza_name != True, f'Please be carefull to writing pizza name...'
add_to = Pizza(x.strip().capitalize())
print(add_to.display_pizza())




request = input("Would you like to add sauces?? (y/n): ")
assert request.lower() in ["y","yes","n","no"] != True, f'Please be carefull for your writing... you can write "yes","y" or "no","n"'



sauces_list = list()
sauces_cost_total = 0
if request.strip().lower() == "y" or request.strip().lower() == "yes":
  while True:
    y= str(input("Which one do you want to add? You can click 'ENTER' if u want to exit. :"))
    y = y.capitalize()    
    if y in [i for i in add_to.description_pizza]:

      # our warning massage if u try to add sauces which already included.
      logger.warning("Its already included. Please add new things without included contents.")
      new_y= str(input("Please Select new sauces without included pizza or you can exit with click 'ENTER' : "))
      new_y = new_y.capitalize()
      sauces_list.append(new_y)
      print("Your order adding pizza....")
      time.sleep(.4)
      print("Your order added your pizza succesfully.")

      if new_y == "":
        break

      # Sauces prices
      new_t=sauces.Sauces_data[new_y]
      sauces_cost_total = new_t + sauces_cost_total

    elif y != [i for i in add_to.description_pizza]:
      if y == "":
         break
      # sauces_order addition
      sauces_list.append(y)
      t=sauces.Sauces_data[y]
      #sauces_order cost
      sauces_cost_total = t + sauces_cost_total
      #-> sauces price = t
      print("Your order adding pizza....")
      time.sleep(.4)
      print("Your order added your pizza succesfully.")

      if y == "":
        break


    #print(AddToPizza.display(adding_to))


  #display our additioned sauces on pizza and its new cost
  adding_to=AddToPizza(sauces,x,add_to.cost_pizza,add_to.description_pizza,sauces_list,sauces_cost_total)
  AddToPizza.definition_new(adding_to)
  AddToPizza.cost_new(adding_to)
  print(AddToPizza.display_pizza(adding_to))

else:
  adding_to=AddToPizza(sauces,x,add_to.cost_pizza,add_to.description_pizza,[],0)
  #display our pizza without any add sauces.
  print(AddToPizza.display_pizza(adding_to))







# USER INFORMATION

print("Please enter your card information.\n")

user_name = input("Name and Surname:")

user_id = input("Your ID:")
try:
  pass
except ValueError:
  print("ID must has only include integer values.")

assert len(user_id) == 11 , f'User ID must has 11 numbers.'

card_number = input("Credit Card No:")
try:
  pass
except ValueError:
  print("Card Number must has only include integer values.")

assert len(card_number) == 16 , f'Card number must has 16 numbers.'

password = input("Credit Card Password:")

order_time = time.strftime("%d-%m-%Y %H:%M")

time.sleep(2)


print("\n")
print("Your order has been received...\n")
print(f"""Your order information:
NAME AND SURNAME: {user_name}
YOUR ORDER: {AddToPizza.display(adding_to)}
ORDER TIME: {order_time}""")
print("\n")
time.sleep(2)

with open("./Orders_Database.csv","a") as file:
  file.write(user_name.capitalize() + ',' + user_id.capitalize() + ',' + card_number + ',' + password + ',' + AddToPizza.display(adding_to) + ',' + order_time + "\n")
encoding_list = ['utf_8', 'utf_8_sig']

### Database username: admin
### Database password: s123

entry = 3

while True:
    data = input("Do you want to view the database?\n")
    if (data == "yes" or data == "YES" or data =="Yes"):
        admin_name = input("Please enter username:")
        admin_password = input("Please enter password:")
        if (admin_name != "admin" and admin_password == "s123"):
            print("Username is wrong...")
            entry -= 1
        elif (admin_name == "admin" and admin_password != "s123"):
            print("Password is wrong...")
            entry -= 1
        elif (admin_name != "admin" and admin_password != "s123"):
            print("Username and password are wrong...")
            entry -= 1
        elif (admin_name == "admin" and admin_password == "s123"):
            print("Successfully logged into the system...")
            with open('Orders_Database.csv','r') as file_read:
              reader = csv.reader(file_read,delimiter = ",")
              for row in reader:
                print(row)
           
               # worked = True
               # try:
                 #   df = pd.read_csv(r"C:\Users\Ascoo\OneDrive\Masaüstü\Python-Course\Intermediate\Orders_Database.csv", encoding=encoding, nrows=6)
                #    print(df)
               # except:
                #    worked = False
               # if worked:
               #     print(encoding, ':\n', df.head())

            break
        if (entry == 0):
            print("You do not have access to the system")
            break
    elif (data == "no" or data == "NO" or data =="No"):
        print("Logging out of the system...")
        break
    else:
        print("You made a wrong choice. Please be careful!")



```
